#data loading and explore

```python
import pandas as pd
```

```python


def try_encodings(file_path):
    encodings = ['utf-8', 'latin1', 'iso-8859-1', 'windows-1252', 'ascii', 'utf-16', 'utf-32']

    for encoding in encodings:
        try:
            df = pd.read_csv(file_path, encoding=encoding)
            print(f"Success encoding: {encoding}")
            return df
        except Exception as e:
            print(f"Failed encoding: {encoding} - Error: {e}")

    print("None of the encodings worked correctly.")
    return None

file_path = 'football_players.csv'

df = try_encodings(file_path)
```

### to see outliers in age and overall

```python
import numpy as np

Q1 = df['Overall'].quantile(0.25)
Q3 = df['Overall'].quantile(0.75)
IQR = Q3 - Q1
upper_bound = Q3 + 1.5 * IQR
lower_bound = Q1 - 1.5 * IQR

outliers = df[(df['Overall'] < lower_bound) | (df['Overall'] > upper_bound)]

print(f"Number of outliers: {len(outliers)}")

```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(x=df['Overall'])
plt.title('Box Plot showing Outliers in Overall')
plt.xlabel('Overall Rating')
plt.show()
```

```python
Q1 = df['Age'].quantile(0.25)
Q3 = df['Age'].quantile(0.75)
IQR = Q3 - Q1
upper_bound = Q3 + 1.5 * IQR
lower_bound = Q1 - 1.5 * IQR

outliers = df[(df['Age'] < lower_bound) | (df['Age'] > upper_bound)]

print(f"Number of outliers: {len(outliers)}")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(x=df['Age'])
plt.title('Box Plot showing Outliers in Age')
plt.xlabel('Age')
plt.show()
```

## 1. Player Overview and Distribution


### What are the most common nationalities represented in the dataset?

```python
df['Nationality'].value_counts().head(10)
```

### What is the distribution of players based on their nationality, age, and overall rating?


```python
df.groupby('Nationality')[['Age', 'Overall']].median().sort_values('Overall', ascending=False).head(20)
```

### Here are some visualizations to help clarify the distribution.

```python
nationality_counts = df['Nationality'].value_counts()

top_20_nationalities = nationality_counts.nlargest(20)

df_top_20 = df[df['Nationality'].isin(top_20_nationalities.index)]


import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.histplot(df_top_20['Nationality'], bins=20, kde=True)
plt.title('Distribution of Players by Nationality')
plt.xlabel('Nationality')
plt.ylabel('Number of Players')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.hist(df['Age'], bins=10, color='lightgreen', edgecolor='black')
plt.title('Distribution of Player Ages')
plt.xlabel('Ages')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Overall'], bins=20, kde=True)
plt.title('Distribution of Overall Ratings')
plt.xlabel('Overall Rating')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Age', y='Overall', data=df)
plt.title('Distribution of Players by Age and Overall Rating')
plt.xlabel('Age')
plt.ylabel('Overall Rating')
plt.show()
```

### Which attributes are most correlated with success in specific positions (e.g., forwards, midfieldersand backwards)?
### OR, How do the preferred positions correlate with players' skill sets? For example, do forwards consistently have higher finishing and shot power?

```python
Forward = df[df['Preferred Positions Type'].str.contains('Forward')]
selected_columns = Forward.iloc[:, 3:15].join(Forward.iloc[:, 20:27])
correlation_matrix = selected_columns.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap for Forward')
plt.show()
```

```python
Midfielder = df[df['Preferred Positions Type'].str.contains('Midfielder')]
selected_columns = Midfielder.iloc[:, 3:15].join(Midfielder.iloc[:, 20:27])
correlation_matrix = selected_columns.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap for Midfielder Positions')
plt.show()
```

```python
Back = df[df['Preferred Positions Type'].str.contains('Back')]
selected_columns = Back.iloc[:, 3:15].join(Back.iloc[:, 20:27])
correlation_matrix = selected_columns.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap for Back Positions')
plt.show()
```

## 5. Comparative Analysis


### How do players from different age groups compare in terms of stamina, sprint speed, and strength?

```python
df.groupby('Age')[['Stamina', 'Sprint speed', 'Strength']].median().sort_values('Stamina', ascending=False)
```

```python
age_stats = df.groupby('Age')[['Stamina', 'Sprint speed', 'Strength']].median()
#Stamina (اللياقة البدنية) Strength (القوة البدنية) Sprint speed (سرعة الجري)
age_stats.plot(kind='bar', figsize=(15, 8))
plt.xlabel("Age")
plt.ylabel("Median Value")
plt.title("Median Stamina, Sprint Speed, and Strength by Age")
plt.legend()
plt.show()
```
